```python
#If-Else Statment
a = 20
b = 30
if b > a:
    print('b is greater than a')

a = 100 
b = 20
if b < a:
    print('a is greater')
else:
    print('b  is greater')

a = 33
b = 33

if b > a:
    print('b is greater than a')
elif a == b:
    print('a and b are equal')
else:
    print('a is greater')

#short Hand if/if_else:
a = 100 
b = 50
if a > b : print('a is greater than b')

a = 50 
b = 60
print('A') if a > b else print('B')

a = 400
b = 400
print('A') if a > b else print ('=') if a == b else print('B') 

#And/Or/Not
a = 100
b = 50 
c = 200
if a > b & b < c:
    print('Condition are true')
else:
    print('condition is not true')

a = 500
b = 100
c = 50 
if a > b or c < b:
    print('Or condition is true')

a = 40
b = 200

if not a > b:
    print('a is not greater than b')

#Nested if 
a = 41

if a > 40:
    print('Above 40')
if a > 22:
    print('also Above 22')
else:
    print('but not above 43')
```

```python
#While loop
i = 1
while i < 6:
    print(i)
    i += 1

i = 2
while i < 10:
    print(i)
    if (i == 8):
        break
    i += 1

i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
   

i = 1
while i < 6:
    print(i)
    i += 1
else:
    print('i is no longer less than 6')
```

```python
#For loop
fruits = ['apple','mango','banana']
for fruit in fruits:
    print(fruit)

for words in "banana":
    print(words)

fruits = ['apple','orange','mango','kiwi']
for fruit in fruits :
    print(fruit)
    if fruit == 'mango':
        break

fruits = ['apple','orange','mango','kiwi']
for fruit in fruits:
    if fruit == 'orange':
        break
    print(fruit)

fruits = ['apple','orange','mango','kiwi']
for fruit in fruits[0:2]:
    print(fruit)

for x in range (0,40,4):
    print(x)

for x in range(6):
    print(x)
else:
    print('Finished')

for x in range(7):
    if x == 5: break
    print(x)
else:
    print('Finally Finished')

color = ['red','yellow','green']
fruits = ['apple','mango','banana']
for colors in color:
    for fruit in fruits:
        print(colors,fruit)
```
